```python
from literalai import LiteralClient
from dotenv import load_dotenv
from typing import Optional, Dict
import os

load_dotenv()
literalai_api_key = os.getenv('LITERAL_API_KEY')

# Initialize the client
literalai_client = LiteralClient(api_key=literalai_api_key)

#Create Users
def create_user_with_name(email: str, first_name: str, last_name: str) -> Dict:
    """
    Create or get a user with email as identifier and first/last name as metadata
    
    Args:
        email (str): User's email address
        first_name (str): User's first name
        last_name (str): User's last name
        
    Returns:
        Dict: User information from Literal AI
    """
    try:
        # Structure the metadata dictionary
        metadata = {
            "first_name": first_name,
            "last_name": last_name
        }
        
        # Use get_or_create_user to either fetch existing user or create new one
        user = literalai_client.api.get_or_create_user(
            identifier=email,
            metadata=metadata
        )
        
        print(f"Successfully processed user: {email}")
        return user
        
    except Exception as e:
        print(f"Error processing user: {str(e)}")
        raise
```

```python
user = create_user_with_name(
        email="jgodbout@colby.edu",
        first_name="Jon",
        last_name="Godbout"
    )
print("User details:", user)
```
